```python
copy=data.copy()
```

```python
copy.shape
```

```python
numerical_columns=data[['credit_score','age','tenure','balance','products_number','estimated_salary']]
numerical_columns
```

```python
catororical_data=data[['country','gender','churn','credit_card','active_member']]
```

```python
catororical_data
```

```python
# Univariate Analysis

plt.figure(figsize=(15, 10))
for idx, col in enumerate(numerical_columns):
    plt.subplot(3, 3, idx + 1)  
    sns.kdeplot(x=col, data=copy, hue="churn", fill=True, alpha=0.6)
    plt.title(f"Distribution of {col.capitalize()} by Churn")
    plt.xlabel(col.capitalize())
    plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for idx, col in enumerate(catororical_data):
    plt.subplot(3, 3, idx + 1)  
    sns.countplot(x=col, data=copy, hue="churn", fill=True, alpha=0.6)
    plt.title(f"Distribution of {col.capitalize()} by Churn")
    plt.xlabel(col.capitalize())
    plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(15, 10))
for idx, col in enumerate(numerical_columns):  
    plt.subplot(3, 3, idx + 1)  
    sns.boxplot(x="churn", y=col, data=copy, palette="Set2")
    plt.title(f"Boxplot of {col.capitalize()} by Churn")
    plt.xlabel("Churn")
    plt.ylabel(col.capitalize())

plt.tight_layout()  # Prevent overlapping
plt.show()
```

```python
corr=copy.drop(['country', 'gender'], axis = 1).corr()
```

```python
corr
```

```python
# Correlation 
plt.figure(figsize= ( 15,10))
sns.heatmap( corr,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
```

```python
# Multivariate Analysis

sns.pairplot(data=copy, hue='churn', palette="Set2", diag_kind="kde")
plt.suptitle("Pairwise Relationships by Churn", y=1.02)
plt.show()
```
